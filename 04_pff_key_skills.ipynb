### `while` Loop

````python
i = 0
while i < 12:
    print(i, end=' ')
    i += 1  # in-place addition
````

````python
i = 0
while i < 12:
    i += 1
    print(i, end=' ')
````

````python
i = 0
while True:
    i += 1
    print(i, end=' ')
    if i > 11:
        break
````

### `if-elif-else` Control Structure

General scheme:

```python
if condition1:
 # block of code
elif condition2:
 # block of code
else:
 # block of code
```

````python
n = 10
````

````python
if n > 5:
    print('Number is greater than 5.')
````

````python
n = 20
````

````python
if n > 5:
    print('Number is greater than 5.')
else:
    print('Number is smaller than 5.')
````

````python
n = 4.76
````

````python
if n > 10:
    print('Number is greater than 10.')
elif n > 5:
    print('Number is greater than 5.')
elif n >= 0:
    print('Number is positive (and smaller than 5).')
else:
    print('Number is negative.')
````

### `try-except-else-finally` Control Structure

General scheme:

```python
try:
# code that might raise an exception
except ExceptionType:
# handle the exception
else:
# code to run if no exception was raised
finally:
# code to run no matter what happens (exception or no exception)
```
   
Specific error:

   ```python
   try:
       # some code
   except ValueError:
       # handle value error
   ```

````python
import math
````

````python
n = -5
````

````python
# math.sqrt(n)
````

````python
try:
    sr = math.sqrt(n)
    print(f'Square root of {n}: {sr}')
except:
    print('Exception.')
    sr = math.sqrt(abs(n))
    print(f'Square root of abs({n}): {sr}')
````

````python
try:
    sr = math.sqrt(n)
    print(f'Square root of {n}: {sr}')
except ValueError:
    print('Exception.')
    sr = math.sqrt(abs(n))
    print(f'Square root of abs({n}): {sr}')
````

````python
if n >= 0:
    sr = math.sqrt(n)
    print(f'Square root of {n}: {sr}')
else:
    sr = math.sqrt(abs(n))
    print(f'Square root of abs({n}): {sr}')
````

````python
n = [1, 2]
````

````python
try:
    sr = math.sqrt(n)
    print(f'Square root of {n}: {sr}')
except ValueError:
    print('Exception.')
    sr = math.sqrt(abs(n))
    print(f'Square root of abs({n}): {sr}')
except TypeError:
    print('Cannot calculate the square root for n.')
    print('n needs to be a real number or integer.')
````
