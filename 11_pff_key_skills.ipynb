### Basics

```python
x = list(range(0, 21, 2))
x
```

```python
y = [i ** 2 for i in x]
y
```

```python
plt.plot(y);
```

```python
plt.plot(x, y);
```

```python
plt.plot(y, x);
```

```python
plt.plot(x, y)
plt.title('y = x ** 2');
```

```python
plt.plot(x, y)
plt.title('$y = x^2$');
```

```python
plt.plot(x, y)
plt.title('$y = x^2$')
plt.xlabel('x')
plt.ylabel('y');
```

```python
plt.plot(x, y, color='green')
plt.title('$y = x^2$')
plt.xlabel('x')
plt.ylabel('y');
```

```python
plt.plot(x, y, linestyle='dashed', color='red')
plt.title('$y = x^2$')
plt.xlabel('x')
plt.ylabel('y');
```

```python
plt.plot(x, y, linestyle='dotted', color='black', marker='o')
plt.title('$y = x^2$')
plt.xlabel('x')
plt.ylabel('y');
```

```python
plt.plot(x, y, linestyle='dotted', color='black', marker='o')
plt.title('$y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True);
```

### Heatmaps

```python
fig, ax = plt.subplots()
i = ax.imshow(rng.random((10, 10)), cmap='Blues')
cbar = ax.figure.colorbar(i)
cbar.ax.set_ylabel('coloring', rotation=-90, va='bottom')
plt.grid(False);
```

```python
c = raw.pct_change().corr()
c.iloc[:6, :6]
```

```python
fig, ax = plt.subplots()
i = ax.imshow(c, cmap='coolwarm')
cbar = ax.figure.colorbar(i)
cbar.ax.set_ylabel('correlation', rotation=-90, va='bottom')
plt.grid(False);
```
