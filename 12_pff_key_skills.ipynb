### Simple Examples

```python
rng = default_rng()
```

```python
a = rng.standard_normal((26, 3))
a[0] = 0.
a = a.cumsum(axis=0)
```

```python
plt.plot(a);
```

```python
df = pd.DataFrame(a)
```

```python
plt.plot(df);
```

```python
df.plot();
```

```python
df.info()
```

```python
df.columns = list('ABC')
```

```python
df.plot();
```

```python
index = pd.date_range('2024-1-1', periods=len(df), freq='B')
index
```

```python
df.index = index
```

```python
df.plot();
```

```python
# help(df.plot)
```

```python
df.plot(title='A Few Random Walks', grid=False);
```

```python
df.plot(title='A Few Random Walks',
       subplots=True, figsize=(4, 7));
```

### Financial Data Visualization

```python
url = 'https://certificate.tpq.io/eod_prices.csv'
```

```python
raw = pd.read_csv(url, index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
raw.plot(figsize=(10, 7));
```

```python
(raw / raw.iloc[0]).plot(cmap='coolwarm');
```

```python
symbols = raw.columns[:2]
```

```python
data = pd.DataFrame(raw[symbols])
```

```python
(data / data.iloc[0]).plot(title='Stock Prices', cmap='coolwarm');
```

```python
rets = data.pct_change()
```

```python
rets.head()
```

```python
# alpha = factor of transparency
rets.plot(alpha=0.7, cmap='coolwarm');
```

```python
rets.plot(kind='hist', bins=55, alpha=0.5, cmap='coolwarm');
```

```python
rets.plot.hist(bins=55, alpha=0.5, cmap='coolwarm');
```

```python
rets.plot.hist(bins=55, cmap='coolwarm', subplots=True);
```

```python
rets.describe()
```

```python
# help(rets.plot.box)
```

```python
rets.plot.box();
```
