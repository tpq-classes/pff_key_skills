### Random Number Generation

#### Pure Python

```python
import random
```

```python
random.random()  # uniformly distributed (pseudo-)random number (between 0 and 1)
```

```python
[random.random() for _ in range(10)]
```

```python
random.gauss(0, 1)  # standard normally distributed random number
```

```python
[random.gauss(0, 1) for _ in range(10)]
```

#### With NumPy

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
rng.random()  # uniformly distributed (pseudo-)random number
```

```python
rng.random(10)
```

```python
rng.random((5, 3))
```

```python
rng.standard_normal()  # standard-normally distributed (pseudo-)random number
```

```python
rng.standard_normal(10)
```

```python
rng.standard_normal((5, 3))
```

#### Speed Comparison

```python
N = 1_000_000
```

```python
%time rnp = [random.gauss(0, 1) for _ in range(N)]
```

```python
%timeit rnp = [random.gauss(0, 1) for _ in range(N)]
```

```python
np.array(rnp).mean()
```

```python
np.array(rnp).std()
```

```python
%time rnn = rng.standard_normal(N)
```

```python
%timeit rnn = rng.standard_normal(N)
```

```python
rnn.mean()
```

```python
rnn.std()
```
