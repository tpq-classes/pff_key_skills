### Combining Functions with Data Structures

```python
l = list(range(10))
l
```

```python
for i in l:
    print(f(i))
```

```python
squares = list()
for i in l:
    squares.append(f(i))
```

```python
squares
```

```python
squares = [f(i) for i in l]
squares
```

```python
def g(l):
    ''' Calculates the sum. '''
    s = 0
    for i in l:
        s += i
    return s
```

```python
g(l)
```

```python
g(squares)
```

```python
sum(l)
```

```python
sum(squares)
```

```python
def cumsum(l):
    ''' Calculates the cumulative sum. '''
    cs = [l[0]]
    for i in range(1, len(l)):
        cs.append(cs[i - 1] + l[i])
    return cs
```

```python
l
```

```python
cumsum(l)
```

```python
cumsum(squares)
```

```python
cumsum((3, 5, 6, 9, 12))
```

### Namespaces

Generally speaking, there is a **local** and a **global namespace**.

```python
y = 100  # global namespace
```

```python
def f():
    y = 20  # local namespace
    return y
```

```python
y  # global
```

```python
f()  # local y is used
```

```python
y  # global
```

```python
x = -1000  # global
```

```python
def f(x):
    # x is now in local namespace
    return x + y  # value of y from global namespace
```

```python
f(10)
```

### Functions within Functions

```python
def f(x):
    def g(x):
        return x ** 2
    return g(x + 1) + 55
```

```python
f(10)
```

```python
def f(x):
    def g(x):
        return x ** 2
    y = x + 1
    return g(y) + 55
```

```python
f(10)
```

```python
def f(x):
    print(f'x={x} for function f()')
    def g(x):
        print(f'x={x} for function g()')
        return x ** 2
    return g(x + 1) + 55
```

```python
f(10)
```

### Positional Arguments

```python
def f(x, y):
    return x ** y
```

```python
f(2, 8)
```

```python
def f(*args):
    return args[0] ** args[1]
```

```python
f(2, 8)
```

```python
# f(2)
```

```python
f(2, 8, 5)
```

```python
def g(*args):
    return args[0] ** args[1] / args[2]
```

```python
g(2, 8, 2)
```

```python
p = (3, 5, 2)  # tuple object as positional arguments
```

```python
g(*p)
```

```python
p_ = [3, 5, 2]  # list object as positional arguments
```

```python
g(*p_)
```

### Keyword Arguments

```python
def f(x, y):
    return x ** y
```

```python
f(2, 8)
```

```python
def f(**kwargs):
    return kwargs['x'] ** kwargs['y']
```

```python
p = {'y': 9, 'x': 2, 'a': 10, 'c': 'python'}
```

```python
f(**p)
```

```python
def f(*args, **kwargs):
    return args[1] ** kwargs['y']
```

```python
ar = (2, 4, 7)
```

```python
kw = {'y': 9, 'x': 2, 'a': 10}
```

```python
f(*ar, **kw)
```

```python
4 ** 9
```

```python
args = (2, 4, 7)
```

```python
kwargs = {'y': 9, 'x': 2, 'a': 10}
```

```python
f(*args, **kwargs)
```
